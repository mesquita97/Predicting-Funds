# Trabalhando com as variações das features

```python
for i in range(0, len(vl_cotas)-1):
    var_cotas = dataset[vl_cotas[i+1]]-dataset[vl_cotas[i]]
var_cotas
```

```python
for i in range(0, len(bolsas)-1):
    if(i<=6):
        var_bovespa = dataset[bolsas[i+1]] - dataset[bolsas[i]]
    else:
        var_down = dataset[bolsas[i+1]] - dataset[bolsas[i]]
var_bovespa
```

```python
var_down
```

Convertendo as células das taxas ipca em float

```python
for i in range(0, len(taxas_ipca)):
     dataset[taxas_ipca[i]] = dataset[taxas_ipca[i]].str.replace(',','.').astype(float)
```

```python
dataset[taxas_ipca]
```

```python
for i in range(0, len(taxas_ipca)-1):
    var_ipca = dataset[taxas_ipca[i+1]]-dataset[taxas_ipca[i]]
var_ipca
```

```python
for i in range(0, len(taxas_selic)-1):
    var_selic = dataset[taxas_selic[i+1]]-dataset[taxas_selic[i]]
var_selic
```

```python
variations = ['var_bovespa', 'var_down', 'var_ipca', 'var_selic']
```

```python
dt_var = pd.DataFrame({"var_cotas": var_cotas, "var_bovespa": var_bovespa, "var_down": var_down,
                       "var_ipca": var_ipca, "var_selic": var_selic})
```

```python
dt_var
```

Plotando os gráficos

```python
sns.pairplot(dt_var,x_vars=variations, y_vars=["var_cotas"], kind="reg", height=5, aspect=.8)
```
