# Trabalhando com as variações das features

```python
var_cotas = []
for i in range(0, len(vl_cotas)-1):
    n = i+1
    globals()['var_cotas_%s' % n] = dataset[vl_cotas[i+1]]-dataset[vl_cotas[i]]
    var_cotas.append(globals()['var_cotas_%s' % n])
```

```python
var_bovespa = []
var_down = []
for i in range(0, len(bolsas)-1):
    n = i+1
    if(i<=6):
        globals()['var_bovespa_%s' % n] = dataset[bolsas[i+1]] - dataset[bolsas[i]]
        var_bovespa.append(globals()['var_bovespa_%s' % n])
    else:
        globals()['var_down_%s' % n] = dataset[bolsas[i+1]] - dataset[bolsas[i]]
        var_down.append(globals()['var_down_%s' % n])
```

```python
for i in range(0, len(taxas_ipca)):
     dataset[taxas_ipca[i]] = dataset[taxas_ipca[i]].str.replace(',','.').astype(float)
```

```python
var_ipca = []
for i in range(0, len(taxas_ipca)-1):
    n = i+1
    globals()['var_ipca_%s' % n] = dataset[taxas_ipca[i+1]]-dataset[taxas_ipca[i]]
    var_ipca.append(globals()['var_ipca_%s' % n])
```

```python
var_selic = []
for i in range(0, len(taxas_selic)-1):
    n = i+1
    globals()['var_selic_%s' % n] = dataset[taxas_selic[i+1]]-dataset[taxas_selic[i]]
    var_selic.append(globals()['var_selic_%s' % n])
```

```python
variations_name = []
for n in range(1, len(vl_cotas)-1):
    variations_name.append('var_bovespa_%s'% n)
    variations_name.append('var_down_%s'% n)
    variations_name.append('var_ipca_%s'% n)
    variations_name.append('var_selic_%s'% n)
variations_name
```

```python
variations = [var_cotas, var_bovespa, var_down, var_ipca, var_selic]
```

```python
dt_var = pd.DataFrame()
```

```python
for i in range(0, len(vl_cotas)-1):
    n = i+1
    dt_var['var_cotas_%s' % n] = var_cotas[i]
    dt_var['var_bovespa_%s'% n] = var_bovespa[i]
    dt_var['var_down_%s' % n] = var_down[i]
    dt_var['var_ipca_%s' % n] = var_ipca[i]
    dt_var['var_selic_%s' %n] = var_selic[i]
```
